```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results=pd.read_csv('covid_indiana.csv')
```

```python
results
```

```python
plt.figure(figsize=(25,10))
for i in range(len(results)):
    plt.plot(results.columns[60:160],results.values[i][60:160],label=results.values[i][1])
plt.xticks(fontsize=2,rotation=90)
plt.legend()
```
